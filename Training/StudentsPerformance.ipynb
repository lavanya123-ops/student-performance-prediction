```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(X_train,y_train)
```

```python
result=LR.predict(X_test)
result
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error
mae=mean_absolute_error(y_test,result)
print("mean_absolute_error:",mae)
mse=mean_squared_error(y_test,result)  
print("MSE",mse)
rmse=root_mean_squared_error(y_test,result)
print("RMSE",rmse)
```

```python
train_score=LR.score(X_train,y_train)
test_score=LR.score(X_train,y_train)
print("LR model Train Score",train_score)
print("LR model Test Score",test_score)
```

```python
from sklearn.tree import DecisionTreeRegressor
DT=DecisionTreeRegressor()
DT.fit(X_train,y_train)
result=DT.predict(X_test)
ae=mean_absolute_error(y_test,result)
mse=mean_squared_error(y_test,result)
rmse=root_mean_squared_error(y_test,result)
print("mean_absolute_error:",mae)
print("MSE",mse)
print("RMSE",rmse)
train_score=DT.score(X_train,y_train)
test_score=DT.score(X_train,y_train)
print("DT model Train Score",train_score)
print("DT model Test Score",test_score)
```

```python
from sklearn.ensemble import RandomForestRegressor
RF=RandomForestRegressor()
RF.fit(X_train,y_train)
result=RF.predict(X_test)
ae=mean_absolute_error(y_test,result)
mse=mean_squared_error(y_test,result)
rmse=root_mean_squared_error(y_test,result)
print("mean_absolute_error:",mae)
print("MSE",mse)
print("RMSE",rmse)
train_score=RF.score(X_train,y_train)
test_score=RF.score(X_train,y_train)
print("RF model Train Score",train_score)
print("RF model Test Score",test_score)
```

```python
#step 1:load all the saved model
import pickle
with open("result_rf_model.pkl","rb")as f:
    rf_model=pickle.load(f)
with open("lb_education.pkl","rb")as f: 
    lb_education=pickle.load(f)
with open("lb_race.pkl","rb")as f:
    lb_race=pickle.load(f)  
    
```

```python
def predict_result(
    gender='male',
    race_ethnicity='group A',
    education='high school',
    math_score=80,
    reading_score=50,
    writing_score=50,
    lunch_free='standard',
    test_preparation='completed'
):
    lst = []

    # Gender encoding (assuming: female=0, male=1)
    if gender == 'female':
        lst.append(0)
    elif gender == 'male':
        lst.append(1)

    # Race/Ethnicity (using trained LabelEncoder)
    race_ethnicity_encoded = lb_race.transform([race_ethnicity])
    lst += list(race_ethnicity_encoded)

    # Education (use transform, not fit_transform)
    education_encoded = lb_education.transform([education])
    lst += list(education_encoded)

    # Numeric scores
    lst.append(math_score)
    lst.append(reading_score)
    lst.append(writing_score)

    # Lunch type (manual one-hot)
    if lunch_free == 'standard':
        lst += [1, 0]
    elif lunch_free == 'free/reduced':
        lst += [0, 1]

    # Test preparation (manual one-hot)
    if test_preparation == 'non':
        lst += [1, 0]
    elif test_preparation == 'completed':
        lst += [0, 1]

    # Prediction
    result = rf_model.predict([lst])
    return result[0]


```
